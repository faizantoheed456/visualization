*Day 3, Topic 1: Annotations & Text*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ax.text() – Placing Free Text
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), color='#3498DB', linewidth=2)

ax.text(5, 0.5, "Peak region", fontsize=12, color='red', 
        ha='center', va='center', fontweight='bold', 
        bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.6))

ax.set_title("Text annotation example", fontweight='bold')
plt.show()
```

```python
#ax.annotate() – Pointing an Arrow to a Data Point
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y, color='#E74C3C', linewidth=2)

ax.annotate('Maximum value', xy=(np.pi/2, 1), xytext=(np.pi/2 + 2, 0.8), 
            arrowprops=dict(arrowstyle='->', color='#2C3E50', lw=1.5), 
            fontsize=12, color='#2C3E50', fontweight='bold')

ax.set_title("Annotation with arrow", fontweight='bold')
plt.show()
```

```python
#Annotating Multiple Points in a Loop
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
years = [2018, 2019, 2020, 2021, 2022]
values = [120, 145, 170, 200, 235]
ax.plot(years, values, marker='o', color='#3498DB', linewidth=2)

for yr, val in zip(years, values):
    ax.annotate(f"${val}K", xy=(yr, val), xytext=(0, 12), textcoords='offset points',
                ha='center', fontsize=9, color='#2C3E50',
                arrowprops=dict(arrowstyle='->', color='#7F8C8D', lw=0.8))

ax.set_title("Revenue Growth with annotations", fontweight='bold')
ax.set_xlabel("Year")
ax.set_ylabel("Revenue ($K)")
plt.show()
```

*Practice Activity – Day 3, Topic 1*
Scenario: A coffee shop tracks quarterly revenue.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [85, 92, 105, 120]    # in thousands of dollars
```

```python
#Create a Figure and Axes with figsize=(8, 5) and dpi=150.
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

#Plot revenue vs. quarters as a line with circle markers, colour '#8E44AD', linewidth 2.5.
ax.plot(quarters, revenue, marker='o', color='#8E44AD', linewidth=2.5)
#Add a title: "Quarterly Revenue – Coffee Shop" — bold, fontsize 14.
ax.set_title("Quarterly Revenue - Coffee Shop", fontweight='bold', fontsize=14)
#Add axis labels: "Quarter" and "Revenue ($K)".
ax.set_xlabel("Quarter")
ax.set_ylabel("Revenue ($K)")

"""Annotate the Q4 point with an arrow pointing to it, the text label "Record: $120K", placed above 
the point using textcoords='offset points', offset (0, 15). Use 
arrowprops with arrowstyle='->', colour '#E74C3C', linewidth 1.5"""
ax.annotate("Record: $120K", xy=('Q4', 120), textcoords='offset points', xytext=(0, 15),
            arrowprops=dict(arrowstyle='->', color='#E74C3C', lw=1.5),
            fontsize=12, color='#2C3E50', fontweight='bold', ha='center')

#Annotate the Q1 point with the text "Start: $85K" using ax.text(), placed at ('Q1', 88) (just below the marker).
ax.text('Q1', 88, "Start: $85K", fontsize=9, color='red',
        ha='center', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

*Day 3, Topic 2: Tick Formatters and Locators*

```python
#Quick Tick Placement with ax.set_xticks() / ax.set_yticks()
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x) * 100, color='#3498DB', linewidth=2)

ax.set_yticks([-100, -50, 0, 50, 100])

ax.set_title("Manually Set Y-ticks", fontweight='bold')
ax.set_xlabel("x")
ax.set_ylabel("Value")
plt.show()
```

```python
#Example: MultipleLocator on y‑axis
import matplotlib.ticker as mticker

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x) * 100, color='#E74C3C', linewidth=2)

ax.yaxis.set_major_locator(mticker.MultipleLocator(25))

ax.set_title("MultipleLocator – Y‑Ticks Every 25", fontweight='bold')
ax.set_xlabel("x")
ax.set_ylabel("Value")
plt.show()
```

```python
#Example: MaxNLocator on x‑axis
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x) * 100, color='#E74C3C', linewidth=2)

ax.xaxis.set_major_locator(mticker.MaxNLocator(10))

ax.set_title("MultipleLocator – X‑Ticks Every 5", fontweight='bold')
ax.set_xlabel("x")
ax.set_ylabel("Value")
plt.show()
```

```python
#Complete Example – Locator + Formatter Together
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

months = np.arange(1, 13)
sales = [120, 135, 148, 162, 175, 190, 205, 220, 215, 230, 240, 260]

ax.plot(months, sales, color='#2ECC71', marker='o', linewidth=2)

ax.xaxis.set_major_locator(mticker.MultipleLocator(2))

ax.yaxis.set_major_locator(mticker.MaxNLocator(8))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'${x/1000:.1f}K'))

ax.set_title("Monthly sales - Formatted Ticks", fontweight='bold', fontsize=14)
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
# Minor Ticks
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

months = np.arange(1, 13)
sales = [120, 135, 148, 162, 175, 190, 205, 220, 215, 230, 240, 260]

ax.plot(months, sales, color='#2ECC71', marker='o', linewidth=2)

ax.xaxis.set_minor_locator(mticker.AutoMinorLocator(2))   # 2 minor ticks between majors
ax.grid(which='minor', linestyle=':', alpha=0.3)

ax.yaxis.set_major_locator(mticker.MaxNLocator(8))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'${x/1000:.1f}K'))

ax.set_title("Monthly sales - Formatted Ticks", fontweight='bold', fontsize=14)
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

*Practice Activity – Day 3, Topic 2*

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
```

```python
days = np.arange(1, 31)
temperatures = np.random.randint(20, 40, size=30)

#Create fig, ax with figsize=(10, 5) and dpi=150.
fig, ax = plt.subplots(figsize=(10, 5), dpi=150)

#Plot days vs. temperatures as a red dashed line with circle markers.
ax.plot(days, temperatures, linestyle='--', marker='o', color='red', linewidth=2)

#Title: "Daily Temperatures – May 2025", bold, fontsize 14.
ax.set_title("Daily Temperatures - May 2025", fontweight='bold', fontsize=14)
ax.set_xlabel("Days")
ax.set_ylabel("Temperature in °C")

#X‑axis: use a MultipleLocator to place ticks every 3 days.
ax.xaxis.set_major_locator(mticker.MultipleLocator(3))

#Y‑axis: use MaxNLocator with 6 ticks, and format them with °C suffix using a FuncFormatter.
ax.yaxis.set_major_locator(mticker.MaxNLocator(6))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'{x}°C'))

#Remove top and right spines.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
