```python
#Using a Colour Map with scatter
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.randint(20, 200, size=100)
values = np.random.rand(100)

sc = ax.scatter(x, y, s=sizes, c=values, cmap='plasma', alpha=0.8, edgecolor='#333333', linewidth=0.3)

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("Intensity (0-1)", fontsize=11)

ax.set_title("Scatter plot with 'Plasma' Color map", fontweight='bold', fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
#Colour Maps with imshow
data = np.random.rand(10, 10)

fig, ax = plt.subplots(figsize=(6, 5), dpi=150)
im = ax.imshow(data, cmap='coolwarm', aspect='auto')
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Value')
ax.set_title("Heatmap with 'coolwarm' Diverging colormap", fontweight='bold')
plt.show()
```

```python
#Creating a Custom Colour Map
from matplotlib.colors import ListedColormap

custom_colors = ['#2C3E50', '#3498DB', '#1ABC9C', '#F1C40F', '#E74C3C']
custom_cmap = ListedColormap(custom_colors, name='my_custom')


np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.randint(20, 200, size=100)
values = np.random.rand(100)
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sc = ax.scatter(x, y, c=values, cmap=custom_cmap, s=sizes, alpha=0.8)
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Custom Colour Scale')
ax.set_title("Scatter with Custom Colour Map", fontweight='bold')
plt.show()
```

*Practice Activity – Day 3, Topic 3*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(2)
x = np.random.rand(80) * 100
y = x * 0.8 + np.random.randn(80) * 10           # correlated data
importance = np.abs(np.random.randn(80)) * 20    # size
score = y - x * 0.7                              # colour mapping
```

```python
#Create fig, ax with figsize=(9, 6) and dpi=150.
fig, ax = plt.subplots(figsize=(9, 6), dpi=150)

#Plot a scatter plot where:
#x and y are the coordinates.
#Size of each point is given by importance.
#Colour of each point is mapped to score using the 'coolwarm' colour map (a diverging map).
#Add alpha=0.8 and edgecolors='#333333' with linewidth=0.3.
sc = ax.scatter(x, y, s=importance, c=score, cmap='coolwarm', linewidth=0.3, edgecolor='#333333', alpha=0.8)

#Add a colour bar and label it 'Score (Y ‑ 0.7·X)'.
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Score (Y - 0.7 * X)')

#Title: "Correlation with Diverging Colour Map" – bold, fontsize 14.
ax.set_title("Correlation with Diverging Color Map", fontweight='bold', fontsize=14)
#X‑label: "X Value". Y‑label: "Y Value".
ax.set_xlabel("X value")
ax.set_ylabel("Y value")

#Remove top and right spines.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Display
plt.show()
```

*Day 3, Topic 4: Style Sheets*

```python
#Viewing Available Styles

import matplotlib.pyplot as plt
import numpy as np
plt.style.available
```

```python
#Applying a Style Globally – plt.style.use()
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), linewidth=2)
ax.plot(x, np.cos(x), linewidth=2)

ax.set_title("ggplot Style", fontweight='bold')
plt.show()
```

```python
#Applying a Style to a Specific Figure Only – style.context()

with plt.style.context('dark_background'):
    fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
    ax.plot(x, np.sin(x), linewidth=2, label='sin')
    ax.plot(x, np.cos(x), linewidth=2, label='cos')
    ax.set_title("Dark background style", fontweight='bold')
    ax.legend()
    plt.show()
```

*Practice Activity – Day 3, Topic 4*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 135, 148, 162, 175, 190]
```

```python
#Use plt.style.use('fivethirtyeight') globally.
plt.style.use('fivethirtyeight')
#Create a Figure and Axes with figsize=(9, 5) and dpi=150.
fig, ax = plt.subplots(figsize=(9, 5), dpi=150)
#Plot a bar chart: months vs. sales, colour '#3498DB', edgecolour 'white', linewidth 1.
ax.bar(months, sales, color='#3498DB', edgecolor='white', linewidth=1)
#Add a title: "Monthly Sales – fivethirtyeight Style" — bold, fontsize 14.
ax.set_title("Monthly Sales - fivethirtyeight Style", fontweight='bold', fontsize=14)
#Add axis labels: "Month" and "Sales ($K)".
ax.set_xlabel("Month")
ax.set_ylabel("Sales ($K)")

plt.show()
#Then: inside a with plt.style.context('dark_background'): block, create a second figure 
#with the same data but as a line plot with markers, and title "Monthly Sales – Dark Mode".
with plt.style.context('dark_background'):
    fig, ax = plt.subplots(figsize=(9, 5), dpi=150)
    ax.plot(months, sales, marker='o', linewidth=2)
    ax.set_title("Monthly Sales - Dark_background Style", fontweight='bold', fontsize=14)
    ax.set_xlabel("Month")
    ax.set_ylabel("Sales ($K)")
    plt.show()
```
