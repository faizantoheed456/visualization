*Day 3, Topic 5: Pie Charts (ax.pie())*

```python
#First Pie Chart
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['Laptop', 'Monitor', 'Keyboard', 'Mouse', 'Headset']
sales = [45, 30, 15, 25, 20]

fig, ax = plt.subplots(figsize=(7, 7), dpi=150)
ax.pie(sales, labels=categories, autopct='%1.1f%%', startangle=90, colors=plt.cm.tab10.colors[:len(categories)])
ax.set_title("Product Sales Distribution", fontweight='bold', fontsize=14)
plt.show()
```

```python
explode = (0, 0.1, 0, 0, 0)

fig, ax = plt.subplots(figsize=(7, 7), dpi=150)

ax.pie(sales, labels=categories, autopct='%1.1f%%', explode=explode, startangle=90, colors=plt.cm.Set3.colors[:len(categories)])
ax.set_title("Product sales - Monitor highlighted", fontweight='bold', fontsize=14)
plt.show()
```

```python
#Custom Slice Edges with wedgeprops
fig, ax = plt.subplots(figsize=(7, 7), dpi=150)
wedges, texts, autotexts = ax.pie(
    sales, labels=categories, autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Pastel1.colors[:len(categories)],
    wedgeprops={'edgecolor':'white', 'linewidth':2},
    pctdistance=0.55
)

for text in autotexts:
    text.set_fontsize(11)
    text.set_fontweight('bold')
    text.set_color('#2C3E50')

for text in texts:
    text.set_fontsize(11)
    text.set_fontweight('bold')
    text.set_color('#2C3E50')
ax.set_title("Product Sales with custom styling", fontweight='bold', fontsize=14)
plt.show()
```

```python
#The Shadow and Start Angle in Practice
fig, ax = plt.subplots(figsize=(7, 7), dpi=150)

ax.pie(sales, labels=categories, autopct='%1.1f%%', startangle=45, shadow=True,
       colors=plt.cm.Set2.colors[:len(categories)])
ax.set_title("Product sales - Shadow & Custom Start Angle", fontweight='bold', fontsize=14)
plt.show()
```

*Practice Activity – Day 3, Topic 5*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['Rent', 'Salaries', 'Marketing', 'Utilities', 'Supplies']
expenses = [5000, 8000, 3000, 1500, 2500]   # in dollars
```

```python
#Create fig, ax with figsize=(8, 8) and dpi=150.
fig, ax = plt.subplots(figsize=(8, 8), dpi=150)

#Plot a pie chart of expenses with labels categories.
#Use autopct='%1.1f%%'.
#Explode the Marketing slice (index 2) with a value of 0.1. All other slices stay at 0.
#Use startangle=90.
#Use the 'Set2' colour map for slice colours.
#Use wedgeprops={'edgecolor': 'white', 'linewidth': 2} for clean slice edges.
explode = (0, 0, 0.1, 0, 0)
wedges, texts, autotexts = ax.pie(expenses, labels=categories, autopct='%1.1f%%', explode=explode, startangle=90, 
       colors=plt.cm.Set2.colors[:len(categories)], wedgeprops={'edgecolor':'white', 'linewidth':2},
      )

#Add a title: "Monthly Business Expenses" — bold, fontsize 14.
ax.set_title("Monthly Business Expenses", fontweight='bold', fontsize=14)

#Style the percentage labels: fontsize 11, bold, dark grey ('#2C3E50').
for text in autotexts:
    text.set_fontweight('bold')
    text.set_color('#2C3E50')
    text.set_fontsize(11)


plt.show()
```

*Day 3, Topic 6: Boxplots and Violin Plots*

```python
#Example: Single Boxplot with Custom Styling
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(1)
data = np.random.normal(70, 15, 200)

fig, ax = plt.subplots(figsize=(6, 5), dpi=150)
bp = ax.boxplot(data, patch_artist=True,
                boxprops=dict(facecolor='#5DADE2', color='#2C3E50', linewidth=1.5),
                medianprops=dict(color='#E74C3C', linewidth=2),
                whiskerprops=dict(color='#2C3E50', linewidth=1.5),
                capprops=dict(color='#2C3E50', linewidth=1.2))

ax.set_title("Single boxplot - Student Scores", fontweight='bold', fontsize=13)
ax.set_ylabel("Score")
ax.set_xticklabels(['Students'])

plt.show()
```

```python
#Example: Multiple Boxplots Side‑by‑Side
np.random.seed(42)
group_a = np.random.normal(65, 12, 100)
group_b = np.random.normal(75, 10, 100)
group_c = np.random.normal(58, 15, 100)

data = [group_a, group_b, group_c]

fig, ax = plt.subplots(figsize=(7, 5), dpi=150)

bp = ax.boxplot(data, patch_artist=True, tick_labels=['Group A', 'Group B', 'Group C'],
                boxprops=dict(facecolor='#3498DB', color='#2C3E50'),
                medianprops=dict(color='#F1C40F', linewidth=2),
                whiskerprops=dict(color='#2C3E50'),
                capprops=dict(color='#2C3E50'))

ax.set_title("Boxplot - Three Groups Compared", fontweight='bold', fontsize=14)
ax.set_ylabel("Score")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
#Violin Plots (ax.violinplot())
fig, ax = plt.subplots(figsize=(6, 5), dpi=150)
violin = ax.violinplot(data[0], positions=[1], showmedians=True, showextrema=True)

for body in violin['bodies']:
    body.set_facecolor('#8E44AD')
    body.set_alpha(0.7)

ax.set_title("Single violin plot", fontweight='bold')
ax.set_ylabel("Value")
ax.set_xticks([1])
ax.set_xticklabels(['Group A'])
plt.show()
```

```python
#Example: Multiple Violins Side‑by‑Side
fig, ax = plt.subplots(figsize=(8, 6), dpi=150)

violins = ax.violinplot(data, showmedians=True, showextrema=True)

colors = ['#3498DB', '#2ECC71', '#F39C12']
for i, body in enumerate(violins['bodies']):
    body.set_facecolor(colors[i])
    body.set_alpha(0.7)

for partname in ('cbars', 'cmaxes', 'cmins', 'cmedians'):
    vp = violins[partname]
    vp.set_edgecolor('#2C3E50')
    vp.set_linewidth(1.2)

ax.set_title("Violin Plots - Three groups Compared", fontweight='bold', fontsize=14)
ax.set_ylabel("Value")
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Group A', 'Group B', 'Group C'])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

*Practice Activity – Day 3, Topic 6*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
class_A = np.random.normal(70, 12, 150)    # Exam scores
class_B = np.random.normal(75, 10, 150)
class_C = np.random.normal(62, 18, 150)
all_scores = [class_A, class_B, class_C]
```

```python
#Create fig, axes with a 1‑row × 2‑column layout, figsize=(14, 6), dpi=150.
fig, axes = plt.subplots(1, 2, figsize=(14, 6), dpi=150)

"""Left panel (axes[0]): Boxplot
All three classes side‑by‑side.
patch_artist=True.
Box face colour '#85C1E9', median colour '#E74C3C', whisker/cap colour '#2C3E50'.
Title: "Boxplot – Exam Scores by Class" (bold, 13).
Y‑label: "Score"
Remove top/right spines."""
bp = axes[0].boxplot(all_scores, tick_labels=['Class A', 'Class B', 'Class C'], patch_artist=True, 
                   boxprops=dict(facecolor='#85C139', color='#2C3E50', linewidth=1.5),
                   medianprops=dict(color='#E74C3C', linewidth=2),
                   whiskerprops=dict(color='#2C3E50', linewidth=1.5),
                   capprops=dict(color='#2C3E50', linewidth=1.5))
axes[0].set_title("Boxplot - Exam Scores by Class", fontweight='bold', fontsize=13)
axes[0].set_ylabel("Score")
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)


"""Right panel (axes[1]): Violin plot
All three classes side‑by‑side.
showmedians=True, showextrema=True.
Customise body colours: '#3498DB', '#2ECC71', '#F39C12' with alpha 0.7.
Title: "Violin Plot – Exam Scores by Class" (bold, 13).
Y‑label: "Score".
Remove top/right spines"""
violin = axes[1].violinplot(all_scores, showmedians=True, showextrema=True)
colors = ['#3498DB', '#2ECC71', '#F39C12']
for i, body in enumerate(violins['bodies']):
    body.set_facecolor(colors[i])
    body.set_alpha(0.7)

for partname in ('cbars', 'cmins', 'cmaxes', 'cmedians'):
    vp = violins[partname]
    vp.set_edgecolor('#2C3E50')
    vp.set_linewidth(1.2)

axes[1].set_title("Violin Plots – Three Groups Compared", fontweight='bold', fontsize=14)
axes[1].set_ylabel("Value")
axes[1].set_xticks([1, 2, 3])
axes[1].set_xticklabels(['Class A', 'Class B', 'Class C'])
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
plt.show()
```
