```python
#Simple Moving Line (Sine Wave Travels)
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib qt
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
x = np.linspace(0, 2 * np.pi, 200)
line, = ax.plot([], [], color='crimson', linewidth=2)   # empty line initially
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)
ax.set_title("Traveling Sine Wave", fontweight='bold')
ax.set_xlabel("x")
ax.set_ylabel("sin(x - shift)")

def update(frame):
    shift = frame * 0.1   # slowly increase the phase shift
    line.set_data(x, np.sin(x - shift))
    return line,          # return the artist(s) that changed

anim = FuncAnimation(fig, update, frames=100, interval=40, repeat=True)
plt.show()
```

```python
#Animated Scatter Plot (One Point Moving)
fig, ax = plt.subplots(figsize=(6, 6), dpi=150)
scat = ax.scatter([], [], s=100, color='teal')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_title("Moving point", fontweight='bold')

def update(frame):
    x = frame % 10 + 0.5
    y = np.sin(frame * 0.3) * 5 + 5
    scat.set_offsets([[x, y]])
    return scat,

anim = FuncAnimation(fig, update, frames=100, interval=50)
plt.show()
```

```python
#Animated Bar Chart (Growing Bars)
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
categories = ['A', 'B', 'C', 'D']
values = np.zeros(4)   # start at zero
bars = ax.bar(categories, values, color='#3498DB', edgecolor='white')
ax.set_ylim(0, 50)
ax.set_title("Growing Bar Chart", fontweight='bold')

def update(frame):
    for bar, h in zip(bars, [10, 20, 15, 25]):
        bar.set_height(min(frame * 0.5, h))   # gradual growth
    return bars

anim = FuncAnimation(fig, update, frames=50, interval=30)
plt.show()
```

```python
#Saving an Animation
anim.save('bar_chart.gif', writer='pillow', fps=25)
```

*Practice Activity – Day 4, Topic 3*
Scenario: Simulate a simple random walk of a single particle in 2D.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# ---------- Simulated Random Walk Data ----------
np.random.seed(42)
steps = 200
x = np.cumsum(np.random.randn(steps) * 0.5)
y = np.cumsum(np.random.randn(steps) * 0.5)

# ---------- Set up the figure and axes ----------
fig, ax = plt.subplots(figsize=(6, 6), dpi=100)
ax.set_xlim(-15, 15)
ax.set_ylim(-15, 15)
ax.set_title("2D Random Walk", fontsize=14, fontweight='bold')
ax.set_xlabel("X")
ax.set_ylabel("Y")

# Draw the entire path as a faint grey line (this stays static)
ax.plot(x, y, color='grey', alpha=0.4, linewidth=1)

# Create an empty scatter that will be moved by the animation
scat = ax.scatter([], [], s=80, c='crimson')

# ---------- Animation Update Function ----------
def update(frame):
    # Set the scatter's position to the current (x[frame], y[frame])
    scat.set_offsets([[x[frame], y[frame]]])
    return scat,

# ---------- Create the Animation ----------
anim = FuncAnimation(fig, update, frames=steps, interval=30, repeat=True)

plt.show()
```
