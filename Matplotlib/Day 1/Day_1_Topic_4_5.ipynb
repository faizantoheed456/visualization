*Day 1, Topic 4: Line Plots (ax.plot())*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Plotting Multiple Lines on the Same Axes
fig, ax = plt.subplots(figsize=(8, 4))
```

```python
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.plot(x, np.sin(x) * 0.5)

fig
```

```python
#A heavily customised line
fig, ax = plt.subplots(figsize=(8, 4))
```

```python
x = np.linspace(0, 10, 20)
ax.plot(x, np.sin(x),
        color='darkorange',
        linewidth=2.5,
        linestyle='-.',
        marker='o',
        markersize=8,
        markerfacecolor='yellow',
        markeredgecolor='black',
        alpha=0.9,
        label='Sine wave'
       )

ax.legend()
fig
```

```python
#Practice Activity
#Scenario: You have two products sold over the same six months.

#Setup
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(9, 5), dpi=150)
```

```python
#Data setup
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_A = [120, 135, 148, 162, 175, 190]
product_B = [210, 195, 200, 188, 172, 160]
```

```python
#Plot product_A vs months as a blue dashed line with circle markers, size 7, linewidth 2.
ax.plot(months, product_A,
        color='blue',
        linestyle='--',
        linewidth=2,
        marker='o',
        markersize=7,
        markerfacecolor='red',
        markeredgecolor='black',
        alpha=0.9,
        label="Product A in months"
       )

#Plot product_B vs months as a red solid line with diamond markers, size 7, linewidth 2.
ax.plot(months, product_B,
        color='red',
        linestyle='-',
        linewidth=2,
        marker='D',
        markersize=7,
        markerfacecolor='yellow',
        markeredgecolor='black',
        alpha=0.9,
        label="Product B in months"
       )

ax.set_title("Product sales comparison", fontsize=15, fontweight='bold')
ax.set_xlabel("Month")
ax.set_ylabel("Units sold")
ax.legend()
fig
```

*Day 1, Topic 5: Scatter Plots (ax.scatter())*

```python
#My first scatter plot
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
```

```python
#Generate random data
np.random.seed(50)
x = np.random.randn(50) * 10 + 50
y = np.random.randn(50) * 15 + 100
```

```python
ax.scatter(x, y)
ax.set_title("Scatter plot of random data")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
fig
```

```python
#Changing Size, Colour, and Transparency
fig, ax = plt.subplots(figsize=(8, 5))
```

```python
x1, y1 = np.random.randn(30) * 5 + 20, np.random.randn(30) * 5 + 30
x2, y2 = np.random.randn(30) * 5 + 50, np.random.randn(30) * 5 + 60
x3, y3 = np.random.randn(30) * 5 + 80, np.random.randn(30) * 5 + 40
```

```python
ax.scatter(x1, y1, s=40, c='red', marker='o', alpha=0.7, label='Group A')
ax.scatter(x2, y2, s=80, c='blue', marker='s', alpha=0.7, label='Group B')
ax.scatter(x3, y3, s=120, c='green', marker='^', alpha=0.7, label='Group C')

ax.legend()
ax.set_title("Three groups - Size and color coded")
fig
```

```python
#The c Parameter and Colour Maps
fig, ax = plt.subplots(figsize=(7, 5))

x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.randint(20, 200, size=100)
values = np.random.rand(100)


sc = plt.scatter(x, y, s=sizes, c=values, cmap='viridis', alpha=0.8, edgecolors='black')

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Color scale (0-1)')

ax.set_title("Scatter with color map and varying sizes")
plt.show()
```

*Practice Activity*
Scenario: You have a dataset of 40 houses with their area (in m²) and price (in thousands of dollars). You also know which houses are in the city centre vs. the suburbs.

```python
import matplotlib.pyplot as plt
import numpy

np.random.seed(1)
area_centre  = np.random.randint(50, 150, size=20)
price_centre = area_centre * 2.5 + np.random.randn(20) * 20
area_suburb  = np.random.randint(80, 200, size=20)
price_suburb = area_suburb * 1.8 + np.random.randn(20) * 15

fig, ax = plt.subplots(figsize=(9, 5), dpi=150)

#Plot the city centre houses as a scatter with marker 'o', colour 'crimson', size 80, and label 'City Centre'.
sc = plt.scatter(area_centre, price_centre, s=80, c='crimson', marker='o', alpha=0.7, label='City Center')
#Plot the suburb houses as a scatter with marker 's', colour 'teal', size 80, and label 'Suburbs'.
sc = plt.scatter(area_suburb, price_suburb, s=80, c='teal', marker='s', alpha=0.7, label='Suburbs')

ax.legend()
ax.set_title("House price vs Area", fontsize=14, fontweight='bold')
ax.set_xlabel("Area (m²)")
ax.set_ylabel("Price (thousands $)")

plt.show()
```
