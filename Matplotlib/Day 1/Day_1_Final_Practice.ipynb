Scenario: A company tracked its advertising spend (in thousands of dollars) and monthly sales (in units sold) for six months.

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
ad_spend = [5, 7, 8, 10, 12, 15]
sales = [120, 135, 150, 160, 180, 200]

#Create a Figure and Axes with figsize=(9,6) and dpi=150.
fig, ax = plt.subplots(figsize=(9, 6), dpi=150)
#Plot a line showing ad_spend vs months — colour '#E67E22', linewidth 3, marker 'o', markersize 8, label 'Ad Spend ($K)'.
ax.plot(months, ad_spend, color='#E67E22', linewidth=3, marker='o', markersize=8, label='Ad Spend ($K)')
#On the same Axes, create a scatter plot of sales vs months — colour '#2ECC71', marker 's', size 120, label 'Sales (Units)'.
ax.scatter(months, sales, c='#2ECC71', marker='s', s=120, label='Sales (Units)')

#Add title: "Monthly Ad Spend vs Sales" — bold, fontsize 15, serif font, colour '#1C2833', pad 15.
ax.set_title("Monthly Ad Spend vs Sales", fontweight='bold', fontsize=15, fontfamily='serif', color='#1C2833', pad=15)
#Add x‑label: "Month" — fontsize 12, colour '#444444'.
ax.set_xlabel("Month", fontsize=12, color='#444444')
#Add y‑label: "Amount / Units" — fontsize 12, colour '#444444'.
ax.set_ylabel("Amount / Units", fontsize=12, color='#444444')

#Use tick_params for labelsize 10 and colour '#666666'.
ax.tick_params(labelsize=10, color='#666666')

#Add legend with loc='upper left', fancybox=True, and title 'Metrics'.
ax.legend(loc='upper left', fancybox=True, title='Metrics')

#Use plt.tight_layout().
plt.tight_layout()

#Save the figure as 'ad_vs_sales.png' (dpi=200, bbox_inches='tight') and 'ad_vs_sales.svg' (bbox_inches='tight').
fig.savefig('ad_vs_sales.png', dpi=200, bbox_inches='tight')
fig.savefig('ad_vs_sales.svg', bbox_inches='tight')

#Display the plot.
plt.show()
```
