```python
#Quick Practice: Plot Temperature Data
#Scenario: You have the daily maximum temperatures in Karachi for the first week of May.

#Setup
import matplotlib.pyplot as plt
import numpy as np

```

```python
#creating a figure and axis
fig, ax = plt.subplots(figsize=(10, 5), dpi=150)
```

```python
#Setting up data
days = np.array(['Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
temperatures = np.random.randint(30, 50, size=(7,))
```

```python
ax.plot(days, temperatures)
#Setting up titles
ax.set_title("Weekly temperatures in Lahore (Summer edition) -May 2026", fontsize=14, fontweight='bold')
ax.set_xlabel("Days Mon-Sun")
ax.set_ylabel("Temperatures (Range: 30-50 °C)")
fig
```
