*Putting It All Together*

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(9, 5), dpi=150)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_A = [120, 135, 148, 162, 175, 190]
product_B = [210, 195, 200, 188, 172, 160]

ax.plot(months, product_A, color='#336633', linewidth=2, marker='o', markersize=7, label='Product A')
ax.plot(months, product_B, color='#33FF66', linewidth=2, marker='s', markersize=7, label='Product B')

ax.set_title("Product sales comparison", fontsize=15, fontweight='bold', fontfamily='serif', color='#FFCC99', pad=15)
ax.set_xlabel("Month", fontsize=12, fontweight='normal', color='#444444', labelpad=10)
ax.set_ylabel("Units sold", fontsize=12, fontweight='normal', color='#444444', labelpad=10)

ax.tick_params(axis='both', labelsize=10, labelcolor='#555555')

ax.legend(loc='upper right', frameon=True, shadow=False, fancybox=True, ncol=1, fontsize=10, title='Products', title_fontsize=11)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
area_centre  = np.random.randint(50, 150, size=20)
price_centre = area_centre * 2.5 + np.random.randn(20) * 20
area_suburb  = np.random.randint(80, 200, size=20)
price_suburb = area_suburb * 1.8 + np.random.randn(20) * 15

#Create a Figure and Axes with figsize=(9,6) and dpi=150.
fig, ax = plt.subplots(figsize=(9, 6), dpi=150)

#Scatter both groups (City Centre and Suburbs) with appropriate colours and labels.
ax.scatter(area_centre, price_centre, c='#0000CC', s=40, marker='o', alpha=0.7, label='City Centre')
ax.scatter(area_suburb, price_suburb, c='#9900CC', s=40, marker='s', alpha=0.7, label='Suburbs')

#Set the title to "House Prices vs. Area" — use a serif font, fontsize 15, bold, dark colour ('#1C2833'), and pad of 15.
ax.set_title("House prices vs. Area", fontfamily='times new roman', fontsize=15, fontweight='bold', color='#1C2833', pad=15)
#Set axis labels to "Area (m²)" and "Price (thousands $)" — fontsize 12, grey colour ('#444444'), labelpad of 10
ax.set_xlabel("Area (m²)", fontsize=12, color='#444444', labelpad=10, fontweight='normal')
ax.set_ylabel("Price (thousands $)", fontsize=12, color='#444444', labelpad=10, fontweight='normal')

#Use tick_params to make tick labels fontsize 10 and colour '#666666'.
ax.tick_params(axis='both', labelsize=10, labelcolor='#666666')

#Add a legend with loc='lower right', fancybox=True, and a legend title 'Location'.
ax.legend(loc='lower right', fancybox=True, shadow=False, frameon=True, ncol=1, fontsize=10, title='Location', title_fontsize=11)

#Call plt.tight_layout() before plt.show().
plt.tight_layout()
plt.show()
```

```python
#Complete Saving Example
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 4), dpi=150)
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), color='crimson', linewidth=2, label='sin(x)')
ax.plot(x, np.cos(x), color='teal', linewidth=2, label='cos(x)')
ax.set_title('Sine and Cosine', fontweight='bold')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.tight_layout()

fig.savefig('trig_functions.png', dpi=200, bbox_inches='tight', transparent=True)
fig.savefig('trig_functions.svg', bbox_inches='tight')

plt.show()
```

Scenario: Take the house price scatter plot you just built in the previous practice, and save it properly.

```python
import matplotlib.pyplot as plt
import numpy

np.random.seed(1)
area_centre  = np.random.randint(50, 150, size=20)
price_centre = area_centre * 2.5 + np.random.randn(20) * 20
area_suburb  = np.random.randint(80, 200, size=20)
price_suburb = area_suburb * 1.8 + np.random.randn(20) * 15

fig, ax = plt.subplots(figsize=(9, 5), dpi=150)

#Plot the city centre houses as a scatter with marker 'o', colour 'crimson', size 80, and label 'City Centre'.
sc = plt.scatter(area_centre, price_centre, s=80, c='crimson', marker='o', alpha=0.7, label='City Center')
#Plot the suburb houses as a scatter with marker 's', colour 'teal', size 80, and label 'Suburbs'.
sc = plt.scatter(area_suburb, price_suburb, s=80, c='teal', marker='s', alpha=0.7, label='Suburbs')

ax.legend()
ax.set_title("House price vs Area", fontsize=14, fontweight='bold')
ax.set_xlabel("Area (m²)")
ax.set_ylabel("Price (thousands $)")

fig.savefig('house_prices.png', dpi=200, bbox_inches='tight', transparent=True)
fig.savefig('house_prices.svg', bbox_inches='tight')
plt.show()
```
