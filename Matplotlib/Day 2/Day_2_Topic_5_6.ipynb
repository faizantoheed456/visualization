```python
#Basic 1×2 Layout: Bar Chart + Histogram
import matplotlib.pyplot as plt
import numpy as np

#Data for the bar chart
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 50]

#Data for the histogram
data = np.random.randn(500) * 15 + 50

fig, axes = plt.subplots(1, 2, figsize=(12, 5), dpi=150)

axes[0].bar(categories, values, color='#3498DB', edgecolor='white', linewidth=1)
axes[0].set_title("Bar chart - sales by category", fontweight='bold')
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Sales")

axes[1].hist(data, bins=20, color='#E74C3C', edgecolor='white', linewidth=0.5)
axes[1].set_title("Histogram - Score distribution", fontweight='bold')
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#2×2 Layout with Mixed Charts
fig, axes = plt.subplots(2, 2, figsize=(10, 8), dpi=150)

x = np.linspace(0, 10, 100)
axes[0, 0].plot(x, np.sin(x), color='#1ABC9C', linewidth=2)
axes[0, 0].set_title("Line: sin(x)", fontweight='bold')
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("sin(x)")

np.random.seed(42)
x_scatter = np.random.rand(50) * 100
y_scatter = np.random.rand(50) * 100
axes[0, 1].scatter(x_scatter, y_scatter, c='#9B59B6', s=50, alpha=0.7)
axes[0, 1].set_title("Scatter: Random points", fontweight='bold')
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("y")


categories = ['P', 'Q', 'R', 'S']
values = [10, 25, 15, 30]
axes[1, 0].bar(categories, values, color='#E67E22', edgecolor='white')
axes[1, 0].set_title("Bar: Category Values", fontweight='bold')
axes[1, 0].set_xlabel("Category")
axes[1, 0].set_ylabel("Value")

data = np.random.randn(300) * 10 + 60
axes[1, 1].hist(data, bins=15, color='#3498DB', edgecolor='white', linewidth=0.5)
axes[1, 1].set_title("Histogram: Distribution", fontweight='bold')
axes[1, 1].set_xlabel("Value")
axes[1, 1].set_ylabel("Frequency")

fig.suptitle("Multiple Chart Types in One Figure", fontsize=16, fontweight='bold', y=1.02)

plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.tight_layout()
plt.show()
```

```python
#Sharing Axes
fig, axes = plt.subplots(2, 1, figsize=(8, 6), dpi=150, sharex=True)

x = np.linspace(0, 10, 100)
axes[0].plot(x, np.sin(x), color='#1ABC9C')
axes[0].set_title("sin(x)")

axes[1].plot(x, np.cos(x), color='#E74C3C')
axes[1].set_title("cos(x)")

plt.tight_layout()
plt.show()
```

*Practice Activity – Day 2, Topic 5*
Scenario: Build a 2×2 dashboard that compares four aspects of a sales dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_A = [120, 135, 148, 162, 175, 190]
product_B = [210, 195, 200, 188, 172, 160]

np.random.seed(1)
scores = np.random.randn(200) * 10 + 50
```

```python
#Create a 2×2 grid with figsize=(12, 10) and dpi=150.
fig, axes = plt.subplots(2, 2, figsize=(12, 10), dpi=150)

#Top‑left: Line chart – both product sales vs. months. Different colours, linewidth 2.5, legend.
axes[0, 0].plot(months, product_A, color='#009999', marker='o', markerfacecolor='darkblue', 
                linewidth=2.5, label='Product A')
axes[0, 0].plot(months, product_B, color='#006699', marker='s', markerfacecolor='darkorange', 
                linewidth=2.5, label='Product B')
axes[0, 0].set_title("Line chart - Product sales vs Months", fontweight='bold', fontsize=14)
axes[0, 0].set_xlabel('Months')
axes[0, 0].set_ylabel("Product sales")

#Top‑right: Bar chart – average sales per product (compute mean of A and B). 
#Two bars, colours '#3498DB' and '#E74C3C', edgecolor 'black'.
colors = ['#3498DB', '#E74C3C']
average = [np.mean(product_A), np.mean(product_B)]
products = ['Products_A', 'Products_B']
axes[0, 1].bar(products, average, color=colors, 
               edgecolor='black', linewidth=0.5)
axes[0, 1].set_title("Bar chart - average sales per product", fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel("Products")
axes[0, 1].set_ylabel("Average sales")

#Bottom‑left: Scatter plot – random scores vs. index (1 to 200). 
#Colour '#2ECC71', marker 'o', size 30, alpha 0.6.
axes[1, 0].scatter(np.arange(200), scores, c='#2ECC71', marker='o', s=30, alpha=0.6)
axes[1, 0].set_title("Scatter plot - random scores vs index (1 to 200)", fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel("Index (1 to 200)")
axes[1, 0].set_ylabel("Scores")

#Bottom‑right: Histogram – scores, bins 15, colour '#F39C12', edgecolor 'white', 
#linewidth 0.5, density=True.
axes[1, 1].hist(scores, bins=15, color='#F39C12', edgecolor='white', linewidth=0.5, density=True)
axes[1, 1].set_title("Histogram - scores", fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel("Value")
axes[1, 1].set_ylabel("Scores")

axes[0, 0].legend()
plt.suptitle("Sales Dashboard - May 2026", fontsize=16, fontweight='bold', y=1.02)
plt.subplots_adjust(wspace=0.3, hspace=0.4)

plt.tight_layout()
plt.show()
```

*Day 2, Topic 6: Sharing Axes & GridSpec*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 6), dpi=150, sharex=True)

x = np.linspace(0, 10, 100)

axes[0].plot(x, np.sin(x), color='#1ABC9C', linewidth=2)
axes[0].set_title("sin(x)", fontweight='bold')
axes[0].set_ylabel("sin(x)")

axes[1].plot(x, np.cos(x), color='#E74C3C', linewidth=2)
axes[1].set_title("cos(x)", fontweight='bold')
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

plt.tight_layout()
plt.show()
```

```python
#Combining Shared Axes with Mixed Chart Types
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), dpi=150, sharex=True)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 135, 148, 162, 175, 190]
profits = [20, 25, 30, 35, 40, 45]

ax1.bar(months, sales, color='#3498DB', edgecolor='white')
ax1.set_title("Monthly sales", fontweight='bold')
ax1.set_ylabel("Sales ($K)")

ax2.plot(months, profits, color='#E74C3C', marker='o', linewidth=2)
ax2.set_title("Monthly profit", fontweight='bold')
ax2.set_xlabel("Month")
ax2.set_ylabel("Profit ($K)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.gridspec as gridspec
```

```python
#GridSpec – Uneven Layouts
#Example: 1 wide + 2 narrow panels
fig = plt.figure(figsize=(10, 8), dpi=150)
gs = gridspec.GridSpec(2, 2, figure=fig, width_ratios=[1, 1], height_ratios=[1, 1])

# Top panel spans both columns
ax_top = fig.add_subplot(gs[0, :])

# Bottom‑left
ax_left = fig.add_subplot(gs[1, 0])

# Bottom‑right
ax_right = fig.add_subplot(gs[1, 1])

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 135, 148, 162, 175, 190]

ax_top.plot(months, sales, color='#3498DB', marker='o', linewidth=2.5, label='Sales')
ax_top.set_title("Monthly Sales (Wide panel)", fontweight='bold')
ax_top.set_ylabel("Sales ($K)")
ax_top.legend()

ax_left.bar(months, sales, color='#2ECC71', edgecolor='white')
ax_left.set_title("Bar chart", fontweight='bold')
ax_left.set_xlabel("Month")
ax_left.set_ylabel("Sales ($K)")

ax_right.hist(np.random.randn(200)*10+50, bins=15, color='#E74C3C', edgecolor='white', linewidth=0.5)
ax_right.set_title("Histogram", fontweight='bold')
ax_right.set_xlabel("Value")
ax_right.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

*Practice Activity – Day 2, Topic 6*
Your task: Build a dashboard with an uneven layout using GridSpec.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_A = [120, 135, 148, 162, 175, 190]
product_B = [210, 195, 200, 188, 172, 160]

np.random.seed(1)
scores = np.random.randn(200) * 10 + 50
```

```python
#Set up a GridSpec with 3 rows and 2 columns using figsize=(10, 12), dpi=150.
fig = plt.figure(figsize=(10, 12), dpi=150)
gs = gridspec.GridSpec(3, 2, figure=fig)

ax_top = fig.add_subplot(gs[0, :])
ax_left = fig.add_subplot(gs[1, 0])
ax_right = fig.add_subplot(gs[1, 1])
ax_last = fig.add_subplot(gs[2, :])

#Top panel (row 0): Span both columns. Show a line chart of both products vs. months with 
#legend and axis labels.
ax_top.plot(months, product_A, color='#99CCFF', linewidth=2, marker='o', markerfacecolor='#66FFFF', 
            markeredgecolor='black', label='Product A')
ax_top.plot(months, product_B, color='#99FFFF', linewidth=2, marker='o', markerfacecolor='#66CCFF', 
            markeredgecolor='black', label='Product B')
ax_top.set_title("Line chart - Products vs Months", fontweight='bold')
ax_top.set_ylabel("Sales ($K)")
ax_top.legend()

#Middle‑left (row 1, col 0): A bar chart of average sales per product (two bars). 
#Colours '#3498DB' and '#E74C3C', edge black.
products = ['Product A', 'Product B']
averages = [np.mean(product_A), np.mean(product_B)]
colors = ['#3498DB', '#E74C3C']
ax_left.bar(products, averages, color=colors, edgecolor='black', linewidth=0.5)
ax_left.set_title("Bar chart - Products vs Average sales", fontweight='bold')
ax_left.set_xlabel("Products")
ax_left.set_ylabel("Average sales")

#Middle‑right (row 1, col 1): A scatter plot of scores vs. 
#their index. Colour '#2ECC71', marker 'o', size 30, alpha 0.6.
ax_right.scatter(np.arange(200), scores, marker='o', s=30, alpha=0.6, c='#2ECC71')
ax_right.set_title("Scatter - Scores vs. Index", fontweight='bold')
ax_right.set_xlabel("Index")
ax_right.set_ylabel("Score")

#Bottom panel (row 2): Span both columns. Show a histogram of scores with 
#bins=20, colour '#F39C12', edge white.
ax_last.hist(scores, bins=20, color='#F39C12', edgecolor='white', linewidth=0.5)
ax_last.set_title("Histogram - Distribution of scores", fontweight='bold')
ax_right.set_xlabel("Score")
ax_right.set_ylabel("Frequency")

plt.suptitle("Dashborad by using GridSpec", fontweight='bold', fontsize=16, y=1.02)

plt.tight_layout()
plt.show()
```
