*Day 2, Topic 3: Customising Bar Charts*

```python
#Error Bars
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['A', 'B', 'C', 'D']
values=[25, 40, 30, 50]
errors=[3, 5, 2, 6]

fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
ax.bar(categories, values, yerr=errors, capsize=5, color='#3498DB', edgecolor='#2C3E50', linewidth=1)
ax.set_title("Bar chart with error bars", fontweight='bold')
ax.set_xlabel("Category")
ax.set_ylabel("Value")
plt.tight_layout()
plt.show()
```

```python
#Stacked Bar Charts
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_A = [30, 40, 35, 50]
product_B = [20, 25, 30, 35]
product_C = [15, 20, 25, 30]
a, b, c = np.array(product_A), np.array(product_B), np.array(product_C)

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

ax.bar(categories, a, color='#3498DB', label='Product A')
ax.bar(categories, b, bottom=a, color='#E74C3C', label='Product B')
ax.bar(categories, c, bottom=a + b, color='#2ECC71', label='Product C')
ax.set_title("Stacked Bar Chart – Sales by Product", fontweight='bold')
ax.set_xlabel("Quarter")
ax.set_ylabel("Total Sales")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
#Professional Styling – Spines and Grid Lines
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
ax.bar(categories, values, color='#48C9B0', edgecolor='white', linewidth=1)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.yaxis.grid(True, linestyle='--', alpha=0.5, color='#CCCCCC')
ax.xaxis.grid(False)

ax.set_title("Clean bar chart", fontweight='bold')
plt.show()
```

*Practice Activity – Day 2, Topic 3*
Scenario: A company tracks quarterly revenue from three regions (North, South, West).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
north = [50, 65, 70, 80]
south = [40, 55, 50, 60]
west  = [30, 35, 45, 50]

# Uncertainties for each region
north_err = [4, 5, 6, 5]
south_err = [3, 4, 4, 3]
west_err  = [2, 3, 3, 4]
```

```python
#Create a Figure and Axes with figsize=(9,6) and dpi=150.
fig, ax = plt.subplots(figsize=(9, 6), dpi=150)

#Build a stacked bar chart using NumPy to simplify the bottom calculations.
#Colour North '#3498DB', South '#E74C3C', West '#2ECC71'.
#Add error bars to each segment (use yerr with capsize=4).
n, s, w = np.array(north), np.array(south), np.array(west)
ax.bar(quarters, n, yerr=north_err, capsize= 4, color='#3498DB', linewidth=1, edgecolor='black', label='North')
ax.bar(quarters, s, yerr=south_err, capsize= 4, bottom=n, color='#E74C3C', linewidth=1, edgecolor='black', label='South')
ax.bar(quarters, w, yerr=west_err, capsize= 4, bottom= n + s, color='#2ECC71', linewidth=1, edgecolor='black', label='West')

#Add a title: "Quarterly Revenue by Region" — bold, fontsize 14, serif font, colour '#1C2833'.
ax.set_title("Quarterly Revenue by Region", fontfamily='serif', fontweight='bold', color='#1C2833', fontsize=14)
#Add axis labels: "Quarter" and "Revenue ($K)".
ax.set_xlabel("Quarter")
ax.set_ylabel("Revenue ($K)")

#Add a legend.
ax.legend()

#Remove top and right spines, add horizontal grid lines (dashed, light grey).
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.5, color='#CCCCCC')
ax.xaxis.grid(False)

#Use plt.tight_layout() and display.
plt.tight_layout()
plt.show()
```

*Day 2, Topic 4: Customising Histograms*

```python
#Explicit Bin Edges for Complete Control
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
data = np.random.randn(500) * 15 + 50

bins = np.arange(0, 110, 10)

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
ax.hist(data, bins=bins, color='#3498DB', edgecolor='white', linewidth=0.5)
ax.set_title("Histogram with exact bin edges", fontweight='bold')
ax.set_xlabel("Score")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Controlling the Range
np.random.seed(420)
data = np.random.randn(500) * 15 + 50



fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
ax.hist(data, bins=20, range=(30, 70), color='#E67E22', edgecolor='white', linewidth=0.5)
ax.set_title("Histogram with exact bin edges", fontweight='bold')
ax.set_xlabel("Score")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
male_data   = np.random.randn(200) * 10 + 55
female_data = np.random.randn(200) * 8 + 60

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
ax.hist([male_data, female_data], bins=20, stacked=True,
        color=['#3498DB', '#E74C3C'], edgecolor='white', linewidth=0.5,
        label=['Male', 'Female'])
ax.set_title("Stacked Histogram – Scores by Gender", fontweight='bold')
ax.set_xlabel("Score")
ax.set_ylabel("Count")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
ax.hist(data, bins=30, cumulative=True, color='#48C9B0', edgecolor='white', linewidth=0.5, density=True)
ax.set_title("Cumulative density histogram", fontweight='bold')
ax.set_xlabel("Score")
ax.set_ylabel("Cumulative density")

plt.tight_layout()
plt.show()
```

```python
#Edge Styles and Transparency
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
ax.hist(data, bins=30, cumulative=True, color='#48C9B0', edgecolor='black', linewidth=0.8, alpha=0.9, density=True)
ax.set_title("Cumulative density histogram", fontweight='bold')
ax.set_xlabel("Score")
ax.set_ylabel("Cumulative density")

plt.tight_layout()
plt.show()
```

*Practice Activity – Day 2, Topic 4*
Scenario: You have customer ages from two stores.

```python
import matplotlib.pyplot as plt
import numpy as np

store_A_ages = np.random.randint(18, 65, size=300)
store_B_ages = np.random.randint(18, 65, size=250)
```

```python
#Create a Figure and Axes with figsize=(9,6) and dpi=150.
fig, ax = plt.subplots(figsize=(9, 6), dpi=150)
bins = np.arange(15, 70, 5)
#Draw a stacked histogram of both stores together.
#Use identical bins: np.arange(15, 70, 5) (15 to 65 in steps of 5).
#Colour Store A '#5DADE2' and Store B '#F1948A'.
#Edge colour white, linewidth 0.6.
ax.hist([store_A_ages, store_B_ages], bins=bins, stacked=True, color=['#5DADE2', '#F1948A'], 
        edgecolor='white', linewidth=0.6, label=['Store A', 'Store B'])
#Add a title: "Stacked Histogram – Customer Ages by Store" — bold, fontsize 14, serif font, colour '#1C2833'.
ax.set_title("Stacked Histogram - Customer Ages by Store", fontweight='bold', fontsize=14, 
             fontfamily='serif', color='#1C2833')
#Add axis labels: "Age" and "Count".
ax.set_xlabel("Age")
ax.set_ylabel("Count")
#Add a legend with labels 'Store A' and 'Store B'.
ax.legend()

#Use plt.tight_layout() and display.
plt.tight_layout()

plt.show()
```
