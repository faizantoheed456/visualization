#  Topic 12: update_layout() Deep Dive

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# title and axis labels
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3], y=[4,5,6], mode='markers'))
fig.update_layout(
    title=dict(text='My Plot Title', font=dict(family='Arial',
                                               size=20,
                                               color='darkblue')),
    xaxis=dict(
        title='X Axis Label',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=True,
        zerolinecolor='darkgray',
        tickmode='linear',
        tick0=0,
        dtick=1,
        tickangle=45,
        tickfont=dict(size=10, color='black'),
        range=[0, 10]
    ),
    yaxis=dict(
        title='Y Axis',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=True,
        zerolinecolor='darkgray',
        tickformat='.2f'       # format tick labels with 2 decimal places
    )
)

fig.show()
```

```python
# font and color

fig.update_layout(
    font=dict(family='Courier New, monospace',
              size=14,
              color='darkblue')
)
```

```python
# background color
fig.update_layout(
    plot_bgcolor='whitesmoke',
    paper_bgcolor='lightgray'
)
```

```python
# legend customization

fig.update_layout(
    legend=dict(
        title='Legend Title',
        orientation='h',
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1,
        bgcolor='rgba(255, 255, 255, 0.7)',
        bordercolor='black',
        borderwidth=1
    )
)
```

```python
# hover mode and interaction
fig.update_layout(
    hovermode='x unified',
    hoverlabel=dict(
        bgcolor='white',
        font_size=14,
        font_family='Arial',
        namelength=-1
    )
)
```

```python
# annotations and shapes

fig.add_annotation(
    x=2, y=5,
    text='Annotation',
    showarrow=True,
    arrowhead=2,
    ax=30, ay=-30,
    font=dict(size=14, color='red')
)
```

```python
# shapes
fig.add_shape(
    type='rect',
    x0=1, x1=3, y0=4, y1=6,
    line=dict(color='blue'),
    fillcolor='rgba(0, 100, 200, 0.3)'
)
```

```python
# full layout example
fig = go.Figure()
fig.update_layout(
    title=dict(text='Custom Layout Example',
               font=dict(size=20, color='darkblue'),
               x=0.5),
    xaxis=dict(title='X Axis', showgrid=True,
               gridcolor='lightgray', zeroline=True),
    yaxis=dict(title='Y Axis', showgrid=True,
               gridcolor='lightgray'),
    plot_bgcolor='whitesmoke',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    legend=dict(title='legend', orientation='v',
                x=0.9, y=0.9),
    hovermode='x unified',
    margin=dict(l=50, r=50, t=50, b=50)
)

fig.show()
```

# Practice Activity

```python
tips = px.data.tips()
```

```python
tips.head()
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(
    x=tips[tips['sex'] == 'Male']['total_bill'],
    y=tips[tips['sex'] == 'Male']['tip'],
    mode='markers',
    marker=dict(size=8, color='violet'),
    name='Male'
))
fig.add_trace(go.Scatter(
    x=tips[tips['sex'] == 'Female']['total_bill'],
    y=tips[tips['sex'] == 'Female']['tip'],
    mode='markers',
    marker=dict(size=8, color='red', symbol='x'),
    name='Female'
))
fig.update_layout(
    title=dict(text='Tips - Full layout customzation', 
               font=dict(size=18, color='darkblue', weight='bold'), 
               x=0.5
              ),
    xaxis=dict(title='Total Bill ($)', 
               showgrid=True, 
               gridcolor='lightgray', zeroline=True,
               tickmode='linear', tick0=0, dtick=10),
    yaxis=dict(title='Tips ($)', 
               showgrid=True, gridcolor='lightgray', 
               tickformat='.1f'),
    plot_bgcolor='whitesmoke',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    legend=dict(title='Gender', orientation='h',
                yanchor='bottom', y=-0.3,
                xanchor='center', x=0.5),
    hovermode='x unified',
    margin=dict(l=50, r=50, t=50, b=50)
)
max_tip_row = tips.loc[tips['tip'].idxmax()]
fig.add_annotation(
    x=max_tip_row['total_bill'],
    y=max_tip_row['tip'],
    text=f"Highest tip: ${max_tip_row['tip']:.2f}",
    showarrow=True,
    arrowhead=2,
    ax=40,
    ay=-40,
    font=dict(size=12, color='darkred')
)
fig.show()
```
