#  Phase 2, Topic 10: go.Figure and go.Scatter

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
# creating an empty figure

go.Figure()
```

```python
# a basic scatter with go.scatter

tips = px.data.tips()

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    mode='markers',
    marker=dict(size=8, color='royalblue', symbol='circle'),
    name='Table Visits'
))

fig.show()
```

```python
# adding a second trace - layering

fig.add_trace(go.Scatter(
    x=tips[tips['time'] == 'Lunch']['total_bill'],
    y=tips[tips['time'] == 'Lunch']['tip'],
    mode='markers',
    marker=dict(size=8, color='orange'),
    name='Lunch'
))
fig.add_trace(go.Scatter(
    x=tips[tips['time'] == 'Dinner']['total_bill'],
    y=tips[tips['time'] == 'Dinner']['tip'],
    mode='markers',
    marker=dict(size=8, color='darkblue', symbol='x'),
    name='Dinner'
))

fig.show()
```

```python
# mode='lines+markers' to combine lines and points
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[1, 2, 3, 4, 5],
    y=[2, 3, 5, 7, 11],
    mode='lines+markers',
    line=dict(color='crimson', width=3, dash='dash'),
    marker=dict(color='gold', size=10, symbol='diamond'),
    name='Prime Numbers'
))

fig.show()
```

```python
# adding text label to points

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[1, 2, 3],
    y=[10, 20, 15],
    mode='markers+text',
    text=['Point A', 'Point B', 'Point C'],
    textposition='top center',
    marker=dict(size=15, color='teal')
))

fig.show()
```

```python
# customisinng hover with hovertext and hoverinfo

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=tips['total_bill'][:10],
    y=tips['tip'][:10],
    mode='markers',
    marker=dict(size=10),
    hovertext=tips['day'][:10],
    hoverinfo='x+y+text'
))

fig.show()
```

```python
# full example
df = px.data.iris()
fig = go.Figure()

# trace 1: scatter for setosa
setosa = df[df['species'] == 'setosa']
fig.add_trace(go.Scatter(
    x=setosa['sepal_length'],
    y=setosa['sepal_width'],
    mode='markers',
    marker=dict(size=10, color='crimson', symbol='circle'),
    name='Setosa'
))

# trace 2 for versicolor
versicolor = df[df['species'] == 'versicolor']
fig.add_trace(go.Scatter(
    x=versicolor['sepal_length'],
    y=versicolor['sepal_width'],
    mode='markers',
    marker=dict(size=10, color='teal', symbol='triangle-up'),
    name='Versicolor'
))

# trace 3 for virginica
virginica = df[df['species'] == 'virginica']
fig.add_trace(go.Scatter(
    x=virginica['sepal_length'],
    y=virginica['sepal_width'],
    mode='markers',
    marker=dict(size=10, color='gold', symbol='square'),
    name='Virginica'
))

# layout
fig.update_layout(
    title='Iris Sepal Measurements - Built with go.Figure',
    xaxis_title='Sepal Length (cm)',
    yaxis_title='Sepal Width (cm)'
)
fig.show()
```

```python
df = px.data.tips()

df.head()
```

```python
fig = go.Figure()

# SCATTER TRACE 
lunch = df[df['time'] == 'Lunch']
fig.add_trace(go.Scatter(
    x=lunch['total_bill'],
    y=lunch['tip'],
    mode='markers',
    marker=dict(size=8, color='darkorange', symbol='circle'),
    name='Lunch'
))

# SCATTER TRACE 
dinner = df[df['time'] == 'Dinner']
fig.add_trace(go.Scatter(
    x=dinner['total_bill'],
    y=dinner['tip'],
    mode='markers',
    marker=dict(size=8, color='royalblue', symbol='x'),
    name='Dinner'
))

fig.update_layout(
    title='Tips - Lunch vs Dinner (Graph Objects)',
    xaxis_title='Total Bill ($)',
    yaxis_title='Tips ($)'
)
fig.show()
```
