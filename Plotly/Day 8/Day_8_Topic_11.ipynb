# Topic 11: go.Bar, go.Histogram, go.Box, go.Violin, and go.Pie

```python
# importing the librariies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# bar chart

fig = go.Figure()
fig.add_trace(go.Bar(
    x=['A', 'B', 'C', 'D'],
    y=[10, 25, 15, 30],
    marker=dict(color=['crimson', 'teal', 'gold', 'royalblue']),
    text=[10, 25, 15, 30],
    textposition='outside',
    name='Values'
))

fig.update_layout(title='Bar Chart with go.Bar')
fig.show()
```

```python
# go.Histogram
tips = px.data.tips()

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=tips['total_bill'],
    nbinsx=20,
    marker=dict(color='orange', line=dict(color='white', width=1)),
    name='Total Bill'
))

fig.update_layout(title='Histogram with go.Histogram')
fig.show()
```

```python
# go.Box

fig = go.Figure()
fig.add_trace(go.Box(
    y=tips[tips['time'] == 'Lunch']['total_bill'],
    name='Lunch',
    marker=dict(color='darkorange')
))
fig.add_trace(go.Box(
    y=tips[tips['time'] == 'Dinner']['total_bill'],
    name='Dinner',
    marker=dict(color='royalblue')
))

fig.update_layout(title='Boxplot with go.Box')

fig.show()
```

```python
# go.Violin
fig = go.Figure()

fig.add_trace(go.Violin(
    x=tips[tips['time'] == 'Lunch']['tip'],
    name='Lunch',
    box_visible=True,
    line_color='darkorange'
))

fig.add_trace(go.Violin(
    x=tips[tips['time'] == 'Dinner']['tip'],
    name='Dinner',
    box_visible=True,
    line_color='royalblue'
))

fig.update_layout(title='Violin Plot with go.Violin')
fig.show()
```

```python
# go.Pie

fig = go.Figure()
fig.add_trace(go.Pie(
    labels=['Thu', 'Fri', 'Sat', 'Sun'],
    values=[100, 150, 200, 180],
    hole=0.4,
    marker=dict(colors=['gold', 'coral', 'royalblue', 'seagreen'])
))

fig.update_layout(title='Donut Chart with go.Pie')

fig.show()
```

```python
fig = go.Figure()
# go.Bar trace
days_avg = tips.groupby('day', as_index=False)['total_bill'].mean()
fig.add_trace(go.Bar(
    x=days_avg['day'],
    y=days_avg['total_bill'],
    name='Average Total Bill',
    marker=dict(color='orangered'),
    texttemplate='%{y:.2f}',
    textposition='outside'
))

# trace distribution
for day in ['Thu', 'Fri', 'Sat', 'Sun']:
    fig.add_trace(go.Box(
        y=tips[tips['day'] == day]['total_bill'],
        x=tips[tips['day'] == day]['day'],
        marker=dict(color='royalblue'),
        name=day,
        showlegend=False
    ))

fig.update_layout(
    title='Average Total Bill and Distribution by Day',
    xaxis_title='Day',
    yaxis_title='Total Bill ($)'
)
fig.show()
```
