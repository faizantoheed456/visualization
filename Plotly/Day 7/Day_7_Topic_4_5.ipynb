# Phase 1, Topic 4: px.bar – interactive bar charts

```python
#Basic bar chart – automatic counting
import plotly.express as px

tips = px.data.tips()

fig = px.bar(tips, x='day')
fig.show()
```

```python
# aggregation with y
# it shows sum of total bill in y axis
fig = px.bar(tips, x='day', y='total_bill')
fig.show()
```

```python
# we can change the aggregation by using histfunc
px.bar(tips, x='day', y='total_bill')
```

```python
#color – grouped bars
px.bar(tips, x='day', y='total_bill', color='sex')
```

```python
#barmode – stacked and relative bars

px.bar(tips, x='day', y='total_bill', color='time', barmode='group')
```

```python
#text_auto data labels on bars

px.bar(tips, x='day', y='total_bill', text_auto=True)
```

```python
# orientation = h for horizontal bar charts
px.bar(tips, x='total_bill', y='day', orientation='h', )
```

```python
#Faceting – facet_col and facet_row
fig = px.bar(tips, x='day', y='total_bill', color='time', facet_col='sex')
fig.show()
```

```python
#full example

px.bar(tips, 
       x='day',
       y='total_bill',
       color='time',
       barmode='group',
       text_auto='.2f',
       title='Total Bill by Day and Time',
       labels={'total_bill':'Total Bill($)', 'day':'Day'},
       color_discrete_sequence=px.colors.qualitative.Set2
      )

```

# Practice Activity

```python
avg_tips = tips.groupby(['day', 'sex'], as_index=False)['tip'].mean()

fig = px.bar(avg_tips, x='day', y='tip', color='sex',
             barmode='group', text_auto='.2f',
             title='Average Tip by Day and Sex',
             labels={'tip': 'Average Tip ($)', 'day': 'Day'})
fig.show()
```

# Phase 1, Topic 5: px.histogram – interactive histograms with faceting, cumulative, and marginal plots

```python
#Basic histogram

import plotly.express as px

tips = px.data.tips()
```

```python
px.histogram(tips, x='total_bill')
```

```python
# controlling bins with nbins

px.histogram(tips, x='total_bill', nbins=20)
```

```python
# color comparing groups
px.histogram(tips, x='total_bill', color='sex')
```

```python
# controlling barmode
px.histogram(tips, x='total_bill', color='time', 
            barmode='overlay', opacity=0.6)
```

```python
#histnorm for normalization

px.histogram(tips, x='total_bill', color='time', 
            histnorm='percent')
```

```python
# normalisation with probabilty
px.histogram(tips, x='total_bill', color='time', 
            histnorm='probability')
```

```python
# normalisation with density
px.histogram(tips, x='total_bill', color='time', 
            histnorm='density')
```

```python
#cumulative 
px.histogram(tips, x='total_bill', color='sex', cumulative=True)
```

```python
#text_auto - to create labels on bars

px.histogram(tips, x='total_bill', nbins=10, 
             text_auto=True)
```

```python
# marginal - to add a side plot on x axis

px.histogram(tips, x='total_bill', marginal='rug')
```

```python
# faceting 

px.histogram(tips, x='total_bill', color='time', 
            facet_col='day')
```

```python
# bivariate histogram

px.histogram(tips, x='total_bill', y='tip', color_discrete_sequence=['#325ea8'])
```

```python
# Full example

px.histogram(
    tips,
    x='total_bill',
    color='time',
    barmode='overlay',
    opacity=0.6,
    nbins=25,
    marginal='box',
    title='Total Box Distribution - Lunch vs Dinner',
    labels={'total_bill': 'Total Bill ($)'}
)
```
