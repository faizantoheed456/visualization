# *Topic 6: px.box and px.violin – distribution comparisons with interactive detail*

```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# load a dataset

df = px.data.tips()

df.head()
```

```python
px.box(df, x='day', y='total_bill')
```

```python
# splitting by color

px.box(df, x='day', y='total_bill', color='time')
```

```python
# points showing inidividual data
# 'all' shows all the data but outliers show only the some statistics
px.box(df, x='day', y='total_bill', color='time', 
      points='all')
```

```python
# used notched for median significance check 
px.box(df, x='day', y='total_bill', notched=True)
```

```python
# faceting by using facet_col/ facet_row

px.box(df, x='day', y='total_bill', color='sex', 
      facet_col='time')
```

# px.violin - full distribution shape

```python
px.violin(df, x='day', y='total_bill')
```

```python
# color, points and facet_col 
px.violin(df, x='day', y='total_bill', color='time', 
         points='all', facet_col='sex')
```

```python
px.violin(df, x='day', y='total_bill', color='time', box=True)
```

```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# load a dataset

df = px.data.tips()

df.head()
```

```python
px.violin(df, x='day', y='tip', color='time',
         points='all', title='Tips by Day and Meal Time',
         labels={'tip':'Tip Amount ($)'})
```

# *Phase 1, Topic 7: px.pie* – interactive pie and donut charts

```python
#aggregated data

day_totals = df.groupby('day', as_index=False)['total_bill'].sum()

day_totals
```

```python
px.pie(day_totals, names='day', values='total_bill') 
```

```python
# donut chart

px.pie(day_totals, names='day', values='total_bill', hole=0.4)
```

```python
# customs colors 
px.pie(day_totals, names='day', values='total_bill',hole=0.4
      ,color_discrete_sequence=px.colors.qualitative.Set2)
```

```python
# hover_data for extra data
px.pie(day_totals, names='day', values='total_bill'
      ,hover_data=['total_bill'])
```

```python
# full example

px.pie(
    df,
    names='day',
    values='total_bill',
    hole=0.3,
    title='Total Bill by Day',
    labels={'total_bill':'Total bill ($)'},
    color_discrete_sequence=px.colors.qualitative.Set3
)
```
