Topic 1: What is Plotly and why you need it.

```python
import plotly.express as px

df = px.data.tips()

fig = px.scatter(df, x='total_bill', y='tip', color='time')
fig.show()
```

Phase 1, Topic 2: px.scatter – parameters, color, size, symbol, and marginal plots

```python
#we can coustimize the color scale
fig = px.scatter(df, x='total_bill', y='tip', color='size', color_continuous_scale='plasma')
fig.show()
```

```python
#size – bubble chart
fig = px.scatter(df, x='total_bill', y='tip', size='size', color='sex')
fig.show()
```

```python
# control size with size_max
fig = px.scatter(df, x='total_bill', y='tip', size='size', size_max=40)
fig.show()
```

```python
# maps categorical columns to marker shapes with symbol
fig = px.scatter(df, x='total_bill', y='tip', symbol='sex', color='sex')
fig.show()
```

```python
#hover_name and hover_data 
fig = px.scatter(df, x='total_bill', y='tip',
                 color='time',
                 hover_name='day',
                 hover_data=['size', 'smoker'])

fig.show()
```

```python
#trendline - auto-filled regression line
fig = px.scatter(df, x='total_bill', y='tip',
                 color='time',
                 hover_name='day',
                 hover_data=['size', 'smoker'],
                 trendline='ols')

fig.show()
```

```python
#regression line by using 'lowess'
fig = px.scatter(df, x='total_bill', y='tip',
                 color='time',
                 hover_name='day',
                 hover_data=['size', 'smoker'],
                 trendline='lowess')

fig.show()
```

```python
#marginal_x and marginal_y – marginal distribution plots
fig = px.scatter(df, x='total_bill', y='tip',
                 marginal_x='violin', color='sex',
                 marginal_y='box')

fig.show()
```

```python
#log_x / log_y – logarithmic axes
fig = px.scatter(df, x='total_bill', y='tip',
                 color='sex', log_y=True
                 )

fig.show()
```

```python
#Full example with nine parameters

fig = px.scatter(
    df, 
    x='total_bill',
    y='tip',
    color='time',
    size='size',
    symbol='sex',
    hover_name='day',
    hover_data=['smoker'],
    trendline='ols',
    marginal_x='histogram',
    title='Tips - Full semantic scatter'
)

fig.show()
```

Phase 1, Topic 3: px.line – interactive line plots with aggregation, confidence bands, and faceting

```python
import plotly.express as px

df = px.data.gapminder()

df.head()
```

```python
#basic line plot

#filtering the one country 
pakistan = df[df['country'] == 'Pakistan']

fig = px.line(pakistan, x='year', y='lifeExp')
fig.show()
```

```python
#color – multiple lines, auto‑aggregation

px.line(df, x='year', y='lifeExp', color='continent', error_y=None)
```

```python
#line_group – grouping without colour

px.line(df, x='year', y='lifeExp', line_group='country')
```

```python
#line_dash – different line styles
fig = px.line(df, x='year', y='lifeExp',
              color='continent', 
              line_dash='continent')

fig.show()
```

```python
#markers – adding data points
px.line(df, x='year', y='lifeExp', color='continent', markers=True)
```

```python
#Faceting – facet_col and facet_row
px.line(df, x='year', y='lifeExp',
        color='country',
        facet_col='continent',
        facet_col_wrap=3)
```

```python
#Full worked example – all semantics combined
fig = px.line(
             df,
             x='year',
             y='lifeExp',
             color='continent',
             line_dash='continent',
             markers=True,
             error_y=None,
             title='Life Expectancy by Continent (1952 - 2007)',
             labels={'lifeExp': 'Life Expectancy (years)', 'year': 'Year'}
)
fig.show()
```

```python
fig = px.line(df, x='year', y='lifeExp', color='continent',
              line_shape='spline')   # smooth curves
fig.show()
```
