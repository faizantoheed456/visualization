```python
#Day 5, Topic 2: Setting the Theme with sns.set_theme()
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#The context parameter – controlling scale
iris = sns.load_dataset('iris')

for ctx in ['paper', 'notebook', 'talk', 'poster']:
    sns.set_theme(context=ctx)
    fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
    sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', ax=ax)
    ax.set_title(f"context={ctx}")
    plt.show()
```

```python
#The style parameter – background and grid
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for sty in styles:
    sns.set_theme(style=sty)
    fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
    sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', ax=ax)
    ax.set_title(f"Style={sty}")
    plt.show()
```

```python
#The palette parameter – default colour palette
palettes = ['deep', 'muted', 'bright', 'dark', 'colorblind']
for pal in palettes:
    sns.set_theme(palette=pal)
    fig, ax = plt.subplots(figsize=(6, 4), dpi=300)
    species_counts = iris['species'].value_counts()
    sns.barplot(x=species_counts.index.astype(str), y=species_counts.values, ax=ax)
    ax.set_title(f"Palettes={pal}")
    plt.show()
```

*Day 5, Topic 3: relplot – the all‑in‑one relational plot*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Basic scatter relplot
sns.set_theme(style='darkgrid')
iris = sns.load_dataset('iris')

sns.relplot(data=iris, x='sepal_length', y='sepal_width')
plt.show()
```

```python
#Adding hue
sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

```python
#Adding size and style
tips = sns.load_dataset('tips')
tips.columns
sns.relplot(data=tips, x='total_bill', y='tip', hue='time', size='size', style='smoker')
plt.show()
```

```python
#Creating subplot grids with col and row
sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species', col='species')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', row='time', col='smoker')
plt.show()
```

```python
#kind='line' – relational line plots
sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species', kind='line', errorbar=None)
plt.show()
```

```python
#Controlling figure size and layout
sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species', col='species', height=4, aspect=1.2)
plt.show()
```

```python
g = sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
g.fig.suptitle("Iris Sepal Measurements", y=1.03)   # add an overall title
g.set_xlabels("Sepal Length (cm)")
g.set_ylabels("Sepal Width (cm)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='darkgrid')
tips = sns.load_dataset('tips')
```

```python
"""Create a scatter plot of total_bill vs tip. Colour the points by time (Lunch/Dinner) and 
size them by size (number of people at the table). Use height=5 and aspect=1.3."""
sns.relplot(data=tips, x='total_bill', y='tip', hue='time', 
            size='size', height=5, aspect=1.3)
plt.show()
```

```python
"""Use kind='line' to show total_bill vs tip as a line. Split by smoker (column panels) and colour by sex. 
Disable confidence intervals with ci=None. Set height=4 and aspect=1."""
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', col='smoker', kind='line', errorbar=None, height=4, aspect=1)
plt.show()
```
