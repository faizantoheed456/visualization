*Day 5, Topic 4: scatterplot – the axes‑level scatter made simple*

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Basic scatter with ax control
sns.set_theme(style='darkgrid')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', ax=ax)
ax.set_title("Iris Sepal Scatter", fontweight='bold')
plt.show()
```

```python
#The hue parameter – colour coding
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1', ax=ax)
plt.show()
```

```python
#The size parameter – sizing markers by a column
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size', ax=ax)
plt.show()
```

```python
#The style parameter – varying marker shape
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='smoker', ax=ax)
plt.show()
```

```python
#Using alpha for transparency
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)

sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', alpha=0.6, ax=ax)
plt.show()
```

```python
#Full example – all semantics combined
fig, ax = plt.subplots(figsize=(8, 6), dpi=150)
sns.scatterplot(data=tips, x='total_bill', y='tip', 
                hue='time', 
                size='size', 
                style='smoker',
                palette='coolwarm',
                sizes=(20, 200),
                alpha=0.8,
                ax=ax)
ax.set_title("Tips - Total bill vs. Tip", fontweight='bold', fontsize=14)
ax.set_xlabel("Total bill ($)")
ax.set_ylabel("Tip ($)")
plt.show()
```

*Day 5, Topic 5: lineplot – the axes‑level line plot*

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Basic line plot – automatic aggregation
sns.set_theme(style='dark')
fmri = sns.load_dataset('fmri')
```

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', ax=ax)
ax.set_title("Avergae FMRI Signal over time", fontweight='bold')
plt.show()
```

```python
#hue – colour coding by group
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', ax=ax)
ax.set_title("fMRI Signal By Event Type", fontweight='bold')
plt.show()
```

```python
#style – varying line dash pattern
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', ax=ax)
ax.set_title("fMRI Signal by Event and Region", fontweight='bold')
plt.show()
```

```python
fmri.head()
```

```python
#units – retaining individual trajectories
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', units='subject', estimator=None, lw=0.5, alpha=0.3, ax=ax)
ax.set_title("Individual subject Trajectories (No aggregation)", fontweight='bold')
plt.show()
```

```python
#ci and err_style – controlling the confidence interval
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', errorbar='sd', err_style='bars', ax=ax)
ax.set_title("Standard Deviation Bars", fontweight='bold')
plt.show()
```

```python
#estimator – changing how values are aggregated
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)

sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', estimator='median', ax=ax)
plt.show()
```

```python
#markers and dashes – visual style control
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', markers=True, dashes=True, ax=ax)
ax.set_title("Lines with Markers, no Dashes", fontweight='bold')
plt.show()
```

```python
#Full example – all concepts combined
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)
sns.lineplot(data=fmri, x='timepoint', y='signal',
             hue='event',
             style='region',
             errorbar='sd',
             err_style='band',
             estimator='mean',
             markers=True,
             ax=ax)

ax.set_title("fMRI Signal - Event x Region", fontweight='bold', fontsize=14)
ax.set_xlabel("Timepoint")
ax.set_ylabel("Signal")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style='darkgrid')
fmri = sns.load_dataset('fmri')
```

```python
"""Create a line plot of signal vs timepoint. Colour lines by event. Use the default 95% 
confidence interval as a band. Set figsize=(8,5) and dpi=150. Title: "fMRI Signal by Event Type"."""

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event')
ax.set_title("fMRI Signal by Event Type", fontweight='bold')
plt.show()
```

```python
"""Create the same plot (signal vs timepoint, hue=event), but change the estimator 
to 'median' and show the confidence as vertical error bars (err_style='bars'). 
    Title: "fMRI Signal – Median with Error Bars"."""

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', estimator='median', err_style='bars', ax=ax)
ax.set_title("fMRI Signal – Median with Error Bars", fontweight='bold')
plt.show()
```

```python
"""Create a line plot of signal vs timepoint for the parietal region only (filter your data first). Use units='subject', 
estimator=None, linewidth 0.6, and alpha=0.4 to show each subject as a faint line. 
Do not use hue. Title: "Individual Subjects – Parietal Region"""
parietal = fmri[fmri['region'] == 'parietal']

fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.lineplot(data=parietal, x='timepoint', y='signal',
             units='subject', estimator=None,
             lw=0.6, alpha=0.4, ax=ax)
ax.set_title("Individual Subjects – Parietal Region", fontweight='bold', fontsize=14)
plt.tight_layout()
plt.show()
```
