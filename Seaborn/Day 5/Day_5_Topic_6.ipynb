```python
#Combining all three
sns.set_theme(style='dark')
tips = sns.load_dataset('tips')
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=150)
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='time',
                size='size',
                style='smoker',
                palette='Set1',
                sizes=(30, 200),
                ax=ax
                )
ax.set_title("Tips - All semantics combined", fontsize=14, fontweight='bold')
ax.set_xlabel("Total Bill ($)")
ax.set_ylabel("Tip ($)")
ax.legend(loc='upper left', frameon=True, fancybox=True)
plt.show()
```

Practice Activity – Day 5, Topic 6
Use the diamonds dataset (built into Seaborn) – it's large, so we'll sample 500 rows for speed:



```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='darkgrid')
diamonds = sns.load_dataset('diamonds').sample(500, random_state=42)
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=300)
sns.scatterplot(data=diamonds, x='carat', y='price', 
                hue='cut', 
                palette='Set2',
                size='table',
                sizes=(20, 200),
                style='color',
                ax=ax
               )

ax.set_title("Diamonds - 5 variables in One Plot", fontsize=14, fontweight='bold')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
